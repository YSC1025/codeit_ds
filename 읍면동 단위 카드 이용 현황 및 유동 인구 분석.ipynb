### 2017년과 2018년 데이터 비교 및 전처리

```python
# df[col].nunique(): 이 부분은 df 데이터프레임에서 현재 열(col)의 고유 값 개수를 계산하는 메서드
# sorted(): 이 배열을 오름차순으로 정렬

def print_unique_values(df):
    object_columns = df.columns[df.dtypes == 'object']
    for col in object_columns:
        print(f'{col} 컬럼의 unique 값 개수: {df[col].nunique()}')
        print(sorted(df[col].unique()), '\n')
```

```python
print_unique_values(jeju_2017)
```

```python
print_unique_values(jeju_2018)
```

### 읍면동명과 업종명 같은 경우에는 값의 개수가 동일하긴 하지만, 
### 정말로 똑같은 값들로 이루어져 있는 게 맞는지 한번 더 확인해 보면 좋을 것 같다

```python
for item in jeju_2017['업종명'].unique():
    if item not in jeju_2018['업종명'].unique():
        print(f'2017년에만 있는 값: {item}')
        
for item in jeju_2018['업종명'].unique():
    if item not in jeju_2017['업종명'].unique():
        print(f'2018년에만 있는 값: {item}')
```

## 데이터 합치기

```python
jeju_2017_2018.head()
```

```python
jeju_pop.head()
```

```python
# 문자열(String) 자료형으로 변환
jeju_pop['연월'] = jeju_pop['연월일'].astype('string')
```

```python
jeju_pop
```

```python
jeju_pop['연월'] = jeju_pop['연월'].str[0:4] + "-" + jeju_pop['연월'].str[4:6]
```

```python
jeju_pop
```

```python
jeju_pop['성별'] = jeju_pop['성별'] + "성"
```

```python
jeju_pop
```

```python
groupby_pop = jeju_pop.groupby(['연월', '시군구명', '읍면동명', '성별']).sum().reset_index()
groupby_pop.head()
```

```python
jeju_df = pd.merge(jeju_2017_2018, groupby_pop, how='left', on=['연월', '시군구명', '읍면동명', '성별'])
jeju_df.head()
```

### *연월일 컬럼 삭제*

```python
jeju_df = jeju_df.drop(columns = '연월일')
```

```python
jeju_df.head()
```

```python
jeju_df['업종명'].unique()
```
